```python
def read_pdf(file_path):
    """Reads a PDF file"""
    doc = fitz.open(file_path)
    return "\n".join([page.get_text() for page in doc])

def read_legal_files(folder_path):
    """Reads all legal files in the folder and returns a dictionary with file names and content."""
    legal_documents = {}

    for file_name in os.listdir(folder_path):
        file_path = os.path.join(folder_path, file_name)
        if file_name.endswith(".pdf"):
            legal_documents[file_name] = read_pdf(file_path)
        else:
            print(f"Skipping unsupported file: {file_name}")

    return legal_documents
```

```python
def get_end_folders(root_folder, skip_folder=".ipynb_checkpoints"):
    end_folders = []

    for dirpath, dirnames, _ in os.walk(root_folder, topdown=True):
        # Remove the folders that should be skipped
        dirnames[:] = [d for d in dirnames if d != skip_folder]

        # If there are no subdirectories left, it's an end folder
        if not dirnames:
            end_folders.append(os.path.join(dirpath, ""))  # Ensure trailing backslash

    return end_folders

# Example usage:
folder_path_read = r"C:\Users\user\Desktop\Legal-Document-Discrepancy-Benchmark-Dataset\full_contract_pdf"
result = get_end_folders(folder_path_read)
print(result)
```

```python
def scan_and_print_non_pdfs(root_folder):
    end_folders = get_end_folders(root_folder)

    for folder in end_folders:
        for file in os.listdir(folder):
            file_path = os.path.join(folder, file)
            if os.path.isfile(file_path) and not file.endswith(".pdf"):
                print(f"Non-PDF File Found: {file_path}")

scan_and_print_non_pdfs(folder_path_read)
```

```python
def rename_files_to_lower(root_folder):
    end_folders = get_end_folders(root_folder)

    for folder in end_folders:
        for file in os.listdir(folder):
            file_path = os.path.join(folder, file)
            lower_file_path = os.path.join(folder, file.lower())

            # Check if it's a file and if renaming is needed
            if os.path.isfile(file_path) and file != file.lower():
                os.rename(file_path, lower_file_path)
                print(f"Renamed: {file_path} -> {lower_file_path}")

# Example usage:
rename_files_to_lower(folder_path_read)
```
